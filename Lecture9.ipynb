```python
# WAP to check if string is pallindrome or not
string = input("Enter a string: ")
string1 = string[::-1]

def pallindrome(string):
  if string == string1:
    print("The string is a pallindrome")
  else:
    print("The string is not a pallindrome")

pallindrome(string)
```

```python
'''
take 2 no. and operator from user
check operator and perform operation
'''

a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
c = input("Enter operator: ")

def calculator():
  if c == "+":
    print(a + b)
  elif c == "-":
    print(a - b)
  elif c == "*":
    print(a * b)
  elif c == "/":
    print(a / b)
  elif c == "%":
    print(a % b)
  else:
    print("Wrong operator")

calculator()
```

```python
'''
WAP to check count of vowel in string
using loop check for each character in string
if character is vowel increment count
'''

s1 = input("Enter a string: ")
s = s1.lower()
def vowel():
  c = 0
  for i in range(len(s)):
    if s[i] == "a" or s[i] == "e" or s[i] == "i" or s[i] == "o" or s[i] == "u":
      c += 1
  print(c)

vowel()
```

```python
string = input("Enter a string: ")

vowels = 'aeiouAEIOU'

def vowel_count():
  count = 0
  for i in string:
    if i in vowels:
      count += 1
  print(count)

vowel_count()
```
