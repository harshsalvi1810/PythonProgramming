```python
'''
Escape characters in python
/t : tab
/n : new line
Letter and spaces are ignored
Collects all matching characters and stores them in a list
'''
sentence = "\t\t\thelloo"
print(sentence)
```

```python
text_to_search = """Regular expressions are extremely useful for matching common patterns"""

# want to search all appearence of 'a'
pattern = re.compile(r'a')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
text_to_search = """\t Regular expressions are \nextremely useful for matching common patterns"""

# want to search all appearence of 'a'
pattern = re.compile(r'\n')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
text_to_search = """2009 Regular expressions are 2009 extremely useful for matching common patterns"""

# want to search all appearence of 'a'
pattern = re.compile(r'2009')
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```
