```python
data = pd.DataFrame({'India':[10,20,30,40],
                     'USA':[40,60,70,80],
                     'UK':[100,200,300,400]})
data
```

```python
data['India'].sum(),data['India'].mean(),data['India'].median(),data['India'].min(),data['India'].max()
```

```python
Q1 = data['India'].quantile(0.25)
Q3 = data['India'].quantile(0.75)

IQR = Q3 - Q1

LW = Q1 - 1.5 * IQR
UW = Q3 + 1.5 * IQR

print(Q1,Q3,IQR,UW,LW)
```

```python
skewness_India = data['India'].skew()
skewness_India
```

```python
kurtosis_India = data['India'].kurtosis()
kurtosis_India
```

```python
import matplotlib.pyplot as plt
data['India'].plot(kind = 'kde')
plt.axvline(25,color = 'r',linestyle = '--') #middle
plt.axvline(32.5,color = 'g',linestyle = '--') #upper bound
plt.axvline(17.5,color = 'g',linestyle = '--') #lower bound
plt.axhline(0.005,color = 'b',linestyle = '--') #bottom
plt.show()
```

```python
df = pd.DataFrame({'Gender':['Male','Female','Male','Female'],
              'Station':['Thane','Kalyan','Bhandup','Diva'],
              'Age':[50,70,90,45],
              "Salary":[1000,3000,5000,7000]})
df
```

```python
male_Data = df[df['Gender'] == 'Male']
male_Data
```

```python
female_Data = df[df['Gender'] == 'Female']
female_Data
```

```python
df['Gender'].value_counts().plot(kind = 'pie',autopct = '%0.2f',explode = [0.02,0.02])
plt.title('Gender Distribution')
plt.show()
```
