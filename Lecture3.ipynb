```python
 # while loop

i = 1

while i == 1:
  print(i)
  i = i + 1
```

```python
 # print all odd number
for i in range(0,21):
  if i % 2 == 1:
    print(i)
```

```python
# print all even number
for i in range(0,21):
  if i % 2 == 0:
    print(i)
```

```python
# string data types along with list

str = ["hello","world","python programming"]

for i in str:
  print(i)
```

```python
str = ("hello","world","python programming")

for i in str:
  print(i)
```

```python
name = "India"

print(type(name))

'''
Note:
When any function is defined inside a class it is called
"Method of respective class"
'''

# Method of Strings

print(name.capitalize()) #Camel case
print(name.upper())
print(name.lower())
print(name.casefold()) #convert ever into 1st letter type
```

```python
name = [1,2,5,7,9]
str = ["a","g","b"]

name.reverse()
print(name)

name.sort()
print(name)

str.sort()
print(str)

print(str.sort())
```
