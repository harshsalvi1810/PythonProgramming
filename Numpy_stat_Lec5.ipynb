```python
'''
Binomial Distribution :

  The binomial distibution is the probability distribution for the number of
  successes in a sequence of Bernoulli trials

  example:
  - Person tossing a coin
    The event to occur Head : p = 0.5 i.e. success
                       Tail : q = (1 - p) i.e. failure
                       note : p + q = 1
'''

import numpy as np
from scipy.stats import binom

n,p = 10,0.5
mean, var, skew, kurtosis = binom.stats(n,p,moments='mvsk')

mean,var,skew,kurtosis
```
