```python
# import data manipulation library
import numpy as np
import pandas as pd

# import data visualization library
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/titanic_dataset - titanic_dataset.csv')

df.head()
```

```python
plt.bar(df['Embarked'],df['Pclass'],color = 'gray')
plt.title('Bar Plot')
plt.xlabel('Embarked')
plt.ylabel('Class')

plt.show()
```

```python
#sns.barplot(df[x = 'Embarked'], df[y = 'Pclass'],color = 'green')
sns.barplot(data = df,x = 'Embarked', y = 'Pclass',color = 'green')
plt.show()
```

```python
year = [2000,2001,2002,2003,2004,2005]
sales_Q1 = [10000,20000,40000,60000,80000,90000]
sales_Q2 = [12000,8000,90000,62000,82000,92000]
```

```python
# ploting Box plot using seaborn library

sns.boxplot(data = sales_Q1)
```

```python
sns.boxenplot(data = sales_Q1)
```

```python
sns.violinplot(data = sales_Q1,color = 'brown')
```

```python
sns.lineplot(x = sales_Q1,y = sales_Q2,linestyle = '-',marker = 'D')
sns.regplot(x = sales_Q1,y = sales_Q2)
```

```python
sns.scatterplot(x = sales_Q1,y = sales_Q2)
```

```python
sns.jointplot(x = sales_Q1,y = sales_Q2)
```

```python
plt.figure(figsize = (10,10))
plt.subplots(2,2)
plt.subplot(2,2,1)
plt.plot(year,sales_Q1)
plt.subplot(2,2,2)
plt.plot(year,sales_Q2)
plt.subplot(2,2,3)
plt.boxplot(x = sales_Q1)
plt.subplot(2,2,4)
plt.boxplot(x = sales_Q2)
plt.show()
```
