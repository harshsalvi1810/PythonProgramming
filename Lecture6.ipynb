```python
'''
define odd and even empty lists
iterate from 1 to 20
if % 2 == 0 then append in even list
else append in odd list
'''

even = []
odd = []

for i in range(0,21):
  if i % 2 == 0 :
    even.append(i)
  else:
    odd.append(i)

print("Even :",even)
print("Odd :",odd)
```
