```python
# OOPS
'''
function assign to variable is object(ex. a = sum()__a is object)
global variable outside function for all code
local variable inside function for only that function
without return() = value + none(data type)
'''
string = input("Enter a string: ")
vowels = 'aeiouAEIOU'

def vowel_count():
  count = 0
  for i in string:
    if i in vowels:
      count += 1
  print(count)

c = vowel_count()
print(c)
```

```python
# Fibonacci less than n

number = int(input("Enter a number: "))

def fib(n):
  a, b = 0, 1
  while a < n:
    print(a,end = " ")
    a, b = b, a+b
  print()

ans = fib(number)
```
